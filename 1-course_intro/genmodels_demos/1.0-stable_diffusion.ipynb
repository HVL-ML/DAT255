```python
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler
```

# Set up model and pipeline

```python
model_id = "stabilityai/stable-diffusion-2"
```

```python
# Use the Euler scheduler here instead
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

# Image generation

```python
prompt = "a photo of a student riding a horse on mars, hq, concept art"
```

```python
images = pipe(prompt, num_images_per_prompt=2)
```

```python
images
```

```python
for image in images.images:
    display.display(image)
```

# A simple application

```python
def predict(prompt):
    image = pipe(prompt)
    return image.images[0]
```

```python
import gradio as gr
```

```python
gr.Interface(fn=predict, inputs="text", outputs=gr.Image(shape=(512, 512))).launch(share=True)
```
